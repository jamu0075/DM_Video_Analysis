Let's see if we can comb through 10 hours of video to find 3 (10 second) clips of varrying speeds and environments to test on our modified object detection algorithm. We only want a rounghly 30 seconds of footage out of 10 hours, but how?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read in our data into DF's
#Added columns to identify the source video
#Added columns to identify video frames
indonesia_df = pd.read_csv('Indonesia_Raw.txt', header=None).iloc[1:]
indonesia_df['Video'] = 'bike'
indonesia_df['Frame'] = range(1, len(indonesia_df) + 1)
nyc_df = pd.read_csv('NYC_Traffic.txt', header=None).iloc[1:]
nyc_df['Video'] = 'car'
nyc_df['Frame'] = range(1, len(nyc_df) + 1)
port_df = pd.read_csv('Port_Nassau.txt', header=None).iloc[1:]
port_df['Video'] = 'boat'
port_df['Frame'] = range(1, len(port_df) + 1)

df = [indonesia_df, nyc_df, port_df]
df = pd.concat(df)
df = df.reset_index(drop=True)

#Uncomment this to see all Million data points
#print(df)

clips = {}
clip_means_list = []

#Parse the video list into 10 second intervals to analyize
for i in range(len(df)):
    if (i%300 == 0) and (i != 0):
        tmp_mean = int(np.mean(df[0][i-300:i]))
        clip_means_list.append(tmp_mean)
        clips[tmp_mean] = (df['Video'][i], df['Frame'][i - 300], df['Frame'][i])
        
#Sort the list for easier calculations
clip_means_list.sort()
    
#Plot all the frame differences
plt.figure(figsize=(7, 8))
plt.hist(clip_means_list, edgecolor='k', alpha=.65)

plt.title('Average Frame Difference For 10 Second Intervals')
plt.ylabel('Frequency')
plt.xlabel('Average Frame Difference Value')
plt.show()
```

This isn't very useful, let's add some important points and adjust the graph

```python
### SECOND ITERATION ###

#Store global mean and index for further calculations
global_mean = int(np.mean(clip_means_list)) 
global_mean_index = min(range(len(clip_means_list)), key=lambda i: abs(clip_means_list[i] - global_mean))

#Find Q1
q1 = int(np.mean(clip_means_list[0:global_mean_index]))

#Find Q3
q3 = int(np.mean(clip_means_list[global_mean_index:]))

#Plot all the frame differences
plt.figure(figsize=(7, 8))
plt.hist(clip_means_list, bins=20, range=((q1 - 5000), (q3 + 5000)), edgecolor='k', alpha=.65)

#Mean line
plt.axvline(global_mean, color='k', linestyle='dashed', linewidth=1)
plt.text(global_mean - 1000, 1200, 'Mean',rotation=90)

#Q1 Line
plt.axvline(q1, color='k', linestyle='dashed', linewidth=1)
plt.text(q1 - 1000, 1200, 'Q1',rotation=90)

#Q3 Line
plt.axvline(q3, color='k', linestyle='dashed', linewidth=1)
plt.text(q3 - 1000, 1200, 'Q3',rotation=90)

plt.title('Average Frame Difference For 10 Second Intervals')
plt.ylabel('Frequency')
plt.xlabel('Average Frame Difference Value')
plt.show()
```

This shows how most of our data (The constant motion bike ride video makes up ~70% of data) is dominating the data. However, we can use Q1 to sample clips from the slower videos (Boat sitting in a harbor). It doesn't matter that there are far less slower videos because we only want these clips to *test*, not teach the algorithm. Therefore, we only need a couple clips and there are thousands to sample from.

Let's find a slow moving clip from Q1

```python
#Find the index of the value closest to Q1 from our list of clip differences
tmp = min(range(len(clip_means_list)), key=lambda i: abs(clip_means_list[i] - q1))

#The value of the 10 second clip difference and its index
#print(clip_means_list[tmp], tmp)

#The dictionary value relating to the clip differnce, gives clip info
print(clips[clip_means_list[tmp]])
```
